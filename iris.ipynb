```python
# create figure and axis
fig, ax=plt.subplots()
# scatter sepal length against sepal width
ax.scatter(df['sepal_length'],df["sepal_width"])

ax.set_title("Iris Dataset")
ax.set_xlabel("sepal length")
ax.set_ylabel("sepal width")
```

```python
df['species'].unique()
```

```python
colors={'Iris-setosa':'r','Iris-versicolor':'b',"Iris-virginica":'g'}
fig,ax=plt.subplots()
for i in range(len(df['sepal_length'])):
  ax.scatter(df["sepal_length"][i],df["sepal_width"][i],color=colors[df['species'][i]])
ax.set_label("Iris dataset")
ax.set_xlabel("sepal length")
ax.set_ylabel('sepal width')
```

```python
# create figure and axis
fig, ax=plt.subplots()
# scatter sepal length against sepal width
ax.scatter(df['petal_length'],df["petal_width"])

ax.set_title("Iris Dataset")
ax.set_xlabel("petal length")
ax.set_ylabel("petal width")
```

```python
colors={'Iris-setosa':'r','Iris-versicolor':'b',"Iris-virginica":'g'}
fig,ax=plt.subplots()
for i in range(len(df["petal_length"])):
  ax.scatter(df["petal_length"][i],df["petal_width"][i],color=colors[df['species'][i]])
ax.set_label("Iris dataset")
ax.set_xlabel("Petal length")
ax.set_ylabel("Petal width")
```

```python
# petal length and petal width has a linear relationship
```

```python
columns=df.columns.drop(["species"])
x_data=range(0,df.shape[0])
fig,ax=plt.subplots()
for column in columns:
  ax.plot(x_data,df[column],label=column)
ax.set_label("Iris Dataset")
ax.legend()
```
